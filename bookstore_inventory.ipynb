```python
import json

new_book = {
    "title": "Atomic Habits",
    "author": "James Clear",
    "price": 14.99,
    "in_stock": True
}

#Task 1 — Read the inventory
#This step loads the JSON file into a Python list
def read_inventory():
    with open("inventory.json", "r") as file:
        inventory = json.load(file)
    print("Total books in inventory:", len(inventory))
    return inventory

#Task 2 — Update and save the inventory
#You append the new book and write the updated list back to the file
# "indent=4" controls how the JSON is formatted when Python writes it to the file,
#and it makes the output human‑readable instead of being one long line.

def update_inventory(inventory, new_book):
    inventory.append(new_book)
    with open("inventory.json", "w") as file:
        json.dump(inventory, file, indent=4)
    return inventory

#Task 3 — Display the updated inventory
#Read the file again and print each book in the required format
def display_inventory():
    with open("inventory.json", "r") as file:
        inventory = json.load(file)

    for book in inventory:
        print(f"Title: {book['title']} | Author: {book['author']} | Price: ${book['price']}")

if __name__ == "__main__":
    inventory = read_inventory()
    updated = update_inventory(inventory, new_book)
    display_inventory()
```
